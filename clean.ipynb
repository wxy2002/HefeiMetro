```python
def deal_data(line1):
    line = []
    for i in line1.split('\n'):
        if i[0].isdigit():
            p = 0
            while i[p].isdigit():
                p += 1
            line.append(i[p:])
        else:
            name = ''
            for j in i:
                if not j.isdigit():
                    name += j
                else:
                    break
            line.append(name)
    return line
```

```python
path = []
for i in range(1, 8):
    p = deal_data(line[i])
    for j in range(len(p) - 1):
        path.append([p[j], p[j + 1], i])
```
